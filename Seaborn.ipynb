It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution.

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, palette="muted", split=True)
```

```python
sns.violinplot(x='sepal_width', y='species', data=data)
```

### pairplot

Pairplot represents pairwise relation across the entire dataframe and supports an additional argument called hue for categorical separation.

What it does basically is create a jointplot between every possible numerical column and takes a while if the dataframe is really huge.

* sns.pairplot(data, hue =, kind ='scatter'/'kde'/'hist'/'reg', diag_kind = 'auto'/'hist'/'kde'/None, *kwargs)

kde: kernel density estimate (KDE) 

we can determine both the kind of plot and the Kind of plot for the diagonal subplots. 

```python
sns.pairplot(data=data, hue='species',diag_kind ='kde')
```

# FacetGrid

Multi-plot grid for plotting conditional relationships.

```python
sns.FacetGrid(data)
```

```python
sns.FacetGrid(data, col="time", row="sex")
```

```python
# To draw a plot on every facet, pass a function and the name of one or more columns 
# in the dataframe to FacetGrid.map():

g = sns.FacetGrid(data, col="time",  row="sex")
g.map(sns.scatterplot, "total_bill","tip")
```

```python
# The variable specification in FacetGrid.map() requires a positional argument mapping, 
# but if the function has a data parameter and accepts named variable assignments, you can also use FacetGrid.map_dataframe():

g = sns.FacetGrid(data, col="time",  row="sex")
g.map_dataframe(sns.histplot, x="total_bill")
```

```python
# The FacetGrid constructor accepts a hue parameter. 
# Setting this will condition the data on another variable and make multiple plots in different colors. 
# Where possible, label information is tracked so that a single legend can be drawn:

g = sns.FacetGrid(data, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

### plt.axvspan() & plt.axhspan()

```python
plt.axvspan(xmin,xmax,ymin,ymax)
# Add a vertical span (rectangle) across the axes.

plt.axhspan(ymin,ymax,xmax,xmin)
# Add a horizontal span
```

```python
data = sns.load_dataset('iris')

sns.scatterplot(x="sepal_length", y="sepal_width",data=data)
plt.axvspan(5.5, 7, alpha=0.5)
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width",data=data)
plt.axhspan(3, 4, alpha=0.5)
```
